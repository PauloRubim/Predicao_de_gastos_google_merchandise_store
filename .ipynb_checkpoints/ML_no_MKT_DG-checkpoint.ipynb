# Predição de gastos na Google Merchandising Store

A ideia aqui é ser capaz de antecipar quanto um usuário vai gastar. Aplicando no Marketing Digital, poderemos, por exemplo, antecipar o retorno de uma campanha e talvez, remodelar suas estratégias, melhorando a performance desta. Além disso, poderemos identificar usuários que estão tendo uma má experiência no seu site e melhorar a experiência dele antes mesmo dele decidir não fazer a compra no seu site. Ou o contrário, identificar o que tem de diferente nos usuários que estão comprando. Vamos também separar cluster de usuários e criar estratégias diferentes para usuários diferentes.  


A base de dados foi criada com Google Analytics no site da Google Merchandising Store, onde os dados são abertos.

## 2.0 Feature engineering 

Aqui vamos criar as variáveis que vão alimentar nosso modelo. Vamos utilizar conhecimentos de negócio para criar variáveis que vão gerar um modelo bom. Como estamos falando de um e-commerce, precisamos de variáveis que vão nos contar sobre o comportamento desse consumidor. Lembrando que o consumidor pode acessar o site várias vezes, ir e voltar, vamos contar um pouco sobre a jornada dele. O primeiro passo que podemos dar é usar variáveis quantitativas, que vão nos mostrar quais usuários interagiram mais.  
Escolhemos quatro variáveis para análise:  
- newVisits, que mostra se é a primeira visita do usuário ou não;
- pageviews, que mostra quantas páginas o usuário viu durante a sessão a partir do momento em que ele acessou o site;
- bounces, que podemos chamar de rejeição. É uma visita que não gera nenhuma interação, é uma visita de apenas uma request;
- hits, que é o volume de interação que o Google recebeu do site. Quanto um usuário interagiu no site.

```python
df.dtypes
```

Podemos perceber que as variáveis quantitativas que escolhemos para o nosso modelo estão como string. Temos que tratar.

```python
quant = ['newVisits', 'pageviews', 'bounces', 'hits', 'transactionRevenue']

for coluna in quant:
    df[coluna] = pd.to_numeric(df[coluna])
    
df.dtypes
```

### Tratando a variável resposta

```python
df.transactionRevenue
```

```python
print(df.transactionRevenue.isnull().sum())
print(df.shape[0])
```

Quase todas as linhas da coluna 'transactionRevenue' são nulas. Dado o contexto do nosso problema, a mioria dos usuários não fazem compras no site, podemos inferir que valores nulos são usuários que não gastaram nada. Podemos transformar esses valores para zero. 

```python
df.transactionRevenue.fillna(0, inplace = True)
df.transactionRevenue
```

```python
df.transactionRevenue.value_counts()
```

Podemos notar que existem valores muito quandes de compra, o que no nosso contexto é um pouco estranho visto que se trata de uma loja de produtos de "baixo custos". Indo na documentação, essa variável tem seu valor multiplicado por 10 a sexta. Então traremos esse valor para o formado real.  

```python
df.transactionRevenue = df.transactionRevenue / 1000000
df.transactionRevenue.value_counts()
```

Sabemos que temos mais visitas do que usuários no site. Precisamos mostrar o que o usuário fez ao longo de sua jornada neste período que estamos analisando. Vamos poder identificar um usuário que tem algum tipo de comportamento e prever quanto ele vai gastar. Precisamos ser capazes de calcular por visitor id o que aconteceu e os resultados desse usuário. Então vamos agrupar os dados por visitor id.

```python
df_quant = df.groupby('fullVisitorId', as_index= False)[quant].sum()
df_quant
```

## 3.0 Treinando o modelo 

```python
y = df_quant['transactionRevenue'].copy()
X = df_quant.drop('transactionRevenue', axis=1)
```

```python
from sklearn.model_selection import train_test_split 

SEED = 42

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = SEED)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_train, y_train)
reg_predict = reg.predict(X_test)
reg_predict
```

### Avaliando os resultados

```python
resultados = pd.DataFrame()
resultados['revenue'] = y_test
resultados['predict'] = reg_predict
resultados['erro'] = reg_predict - y_test
```

```python
resultados.head()
```

```python
# vendo os valores previstos para os usuários que gastaram
resultados[resultados.revenue > 0]
```

Percebe-se que o erro está muito grande, o que nos leva a crer que não temos um bom modelo. Podemos também utilizar algumas métricas de avalização. 

```python
import numpy as np 
from sklearn.metrics import mean_squared_error
```

### Mean Square Error (MSE)

```python
print(np.mean((reg_predict - y_test)**2)) # média do erro ao quadrado
print(mean_squared_error(y_test,reg_predict))
```

### Root Mean Square Error (RMSE)

```python
print(np.sqrt(np.mean((reg_predict - y_test)**2))) # raiz do MSE
print(np.sqrt(mean_squared_error(y_test,reg_predict)))
```

Como saber se nosso erro está dentro do aceitavel?

```python
np.mean(df_quant.transactionRevenue)
```

```python
np.std(df_quant.transactionRevenue)
```

Comparando as medias e desvios padrão entre a nossa base e os erros das predições nota-se que os erros estão além do desvio padrão natural da nossa base. O que nos levar a crer que o nosso modelo não está bom. 

```python
import seaborn as sns
```

```python
sns.boxplot(reg_predict)
```

```python
sns.boxplot(y_test)
```

Comparando os dois últimos gráficos vemos uma grande concentração no zero e vários outlinners. O valor realizado está bem diferente do que o modelo previu.

```python
sns.distplot(resultados.erro)
```

Neste histograma que mostra a distribuição do nosso erro. Ele mostra que temos uma base desbalanceada, com vários usuários que não compraram e alguns que compraram. Precisamos inserir mais inteligência no modelo, através de novas variáveis, treinar novos modelos, mexer nos algoritmos que estamos utilizando, para criar um modelo muito bom.

## 4.0 Melhorando o feature engineering

Nessa nossa primeira versão do modelo utilizamos somente algumas variáveis, todas quantitativas. A nossa base de dados é bem mais rica incluindo também variáveis qualitativas. Seria interessante utiliza-las no nosso modelo também. Talvez seja interessante ver informação sobre a primeira e a última visita de um usuário, para entender o primeiro device, o último. Se ele veio de uma mídia orgânica, se foi impactado. Esse tipo de informação costuma ser bem rica para alimentar nosso modelo. Vamos considerar para cada usuário os valores quantitativos e qualitativos da primeira e última visita.

```python
# aqui estamos buscando qual a última visita de um usuário
visitas_ultima = df.groupby('fullVisitorId',as_index=False)['visitNumber'].max()
visitas_ultima
```

Faremos a concatenação desse dataframe com as informações do dataframe principal.

```python
# removendo as linhas duplicadas para fazermos a concatenação. 
usuarios_visitas_unicos = df.drop_duplicates(subset=['fullVisitorId','visitNumber'])
```

```python
visitas = pd.merge(visitas_ultima, usuarios_visitas_unicos, left_on=['fullVisitorId','visitNumber'],
                                                            right_on=['fullVisitorId','visitNumber'], how='left')
visitas
```

```python
# Aqui estamos buscando qual a primeira visita de um usuário, como esta base de dados é apenas uma amostra então poder que não seja 1 (um) 
visitas_primeira = df.groupby('fullVisitorId',as_index=False)['visitNumber'].min()
visitas_primeira
```

```python
# Preciso fazer essa transformação para poder fazer o join 
visitas_primeira.set_index('fullVisitorId',inplace=True)
visitas_primeira
```

```python
visitas = visitas.join(visitas_primeira,how='left',on='fullVisitorId',rsuffix='_primeira')
visitas
```

```python
visitas = pd.merge(visitas,usuarios_visitas_unicos,left_on=['fullVisitorId','visitNumber_primeira'],
                  right_on=['fullVisitorId','visitNumber'],how='left', suffixes=['_ultima','_primeira'])

visitas
```

```python
visitas.head()
```

Existem variáveis quantitativas que incluímos na nossa regra de primeira e última visita. Tínhamos definido outra estratégia lá em cima, que é fazer a soma de cada visita do usuário. Não precisamos dessas variáveis como primeira e última visita. Vamos tirar.

```python
quant
```

```python
for coluna in quant:
    visitas.drop(coluna + '_ultima',axis=1,inplace=True)
    visitas.drop(coluna + '_primeira',axis=1,inplace=True)
```

Nosso dataset tem algumas variáveis do formato id. Essas informações são importantes para fazer transformações. É importante notar que elas não vão ser importantes para o nosso modelo. Elas tendem a identificar uma chave, um valor único. Não vai trazer informação relevante. Como seu id pode mostrar para o modelo qual vai ser o comportamento no site? Não faz muito sentido. Precisamos de informações que falem do usuário, do processo de navegação, sobre o comportamento, e não um id gerado de forma aleatória. Vamos retirar essas variáveis também. Deixaremos o 'fullVisitorId' pois ele é nossa chave principal e talvez precisamos dele para mais transformações. 

```python
ids = ['sessionId_ultima','visitId_ultima','sessionId_primeira', 'visitId_primeira']

visitas.drop(ids,axis=1,inplace=True)

visitas.head()
```

Dentro do dataset, temos várias informações do tipo geográfica. Cidades, região, continente, país. Se pensarmos do ponto de vista do usuário, dificilmente vai mudar uma informação dessa entre várias visitas. Então, para informações regionais, vamos manter apenas o último lugar em que esse usuário acessou o Google.

```python
visitas.columns
```

```python
geo = ['city_primeira','continent_primeira','country_primeira','metro_primeira','region_primeira',
      'networkDomain_primeira','subContinent_primeira']

visitas.drop(geo,axis=1,inplace=True)
```

### Criando variáveis diversas

Vamos trazer de volta o df_quant para o visitas. Assim, teremos informações sobre a primeira e última visita do usuário e o total das informações quantitativas ao longo de todas as visitas.

```python
visitas.shape
```

```python
visitas = pd.merge(visitas,df_quant,left_on=['fullVisitorId'],
                  right_on=['fullVisitorId'],how='left')

visitas.head()
```

A coluna visit start time nos trás uma informação bem valiosa. Ela traz o momento exato da visita em time stamp. É uma conversão número inteiro que vai mudando de segundo em segundo conforme um relógio que começou muitos anos atrás. Cada segundo é um número único. É muito utilizado para representar date time, para trabalhar com data e hora. Podemos através dessa informação saber quanto tempo o usuário demorou entre primeiro e último acesso. Pode ser uma informação bem rica para o nosso modelo.

```python
visitas['tempo_dif'] = visitas.visitStartTime_ultima - visitas.visitStartTime_primeira
```

Não temos nenhuma coluna que nos mostre quantas visitais um usuário fez. Podemos cria-la.

```python
visits = df.groupby('fullVisitorId',as_index=False).count().visitNumber.values
visitas['visits'] = visits
visitas.head()
```

Em relação a coluna do tipo date, poderíamos extrair informações dessa variável para saber o ano, o dia, em que o usuário fez o acesso. O mês que o usuário acessa pode fazer diferença, pela sazonalidade, um ano pode ser melhor que outro.

```python
data = '20160904'

print(data[0:4])
print(data[4:6])
print(data[6:8])
```

```python
visitas['ano_ultima'] = pd.to_numeric([data[0:4] for data in visitas.date_ultima])
visitas['mes_ultima'] = pd.to_numeric([data[4:6] for data in visitas.date_ultima])
visitas['dia_ultima'] = pd.to_numeric([data[6:8] for data in visitas.date_ultima])

visitas['ano_primeira'] = pd.to_numeric([data[0:4] for data in visitas.date_primeira])
visitas['mes_primeira'] = pd.to_numeric([data[4:6] for data in visitas.date_primeira])
visitas['dia_primeira'] = pd.to_numeric([data[6:8] for data in visitas.date_primeira])
```

### Aplicando encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Aqui garantimos que todos os valos de tal coluna são texto, se não for, é feita a conversão 
strings = list(X.operatingSystem_ultima.values.astype('str'))
```

```python
lbl = LabelEncoder()
```

```python
lbl.fit(strings)
```

```python
lbl.transform(strings)
```

```python
# Replicando a ideia para todas as colunas 
for coluna in quali:
    lbl = LabelEncoder()
    strings = list(X[coluna].values.astype('str'))
    lbl.fit(strings)
    X[coluna] = lbl.transform(strings)
```

```python
X.head()
```

### Regressão linear

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
reg_predict = reg.predict(X_test)
```

```python
resultados = pd.DataFrame()
resultados['revenue'] = y_test
resultados['predict'] = reg_predict
resultados['erro'] = reg_predict - y_test
```

```python
resultados.head()
```

Vemos que nosso modelo não está mais preso, mas temos valores negativos. Dentro do comportamento padrão, não temos usuários gastando valores negativos. Para isso, vamos converter todos os valores negativos para zero.

```python
reg_predict[reg_predict < 0] = 0
resultados = pd.DataFrame()
resultados['revenue'] = y_test
resultados['predict'] = reg_predict
resultados['erro'] = reg_predict - y_test
resultados.head()
```

```python
resultados[resultados.revenue > 0]
```

Vendo as variáveis dentro de resultados que os usuários gastaram alguma coisa, ou seja, maior do que zero, saímos do comportamento padrão e temos um modelo em que os usuários gastam mais. Em alguns chegamos bem perto do previsto. Este modelo já melhorou bastante em relação ao primeiro.

```python
np.sqrt(mean_squared_error(y_test,reg_predict))
```

Com este resultado já estamos errando dentro do desvio padrão natural dos dados, mais um indicativo de melhora do nosso modelo.

```python
sns.boxplot(reg_predict)
```

E o boxplot também já se aproximou mais, visualmente, da nossa base de dados. Agora vamos testar se trocando o algoritmo conseguimos melhorar ainda mais o modelo. Para este teste iremos usar o gradiente boosting.

### Treinando um gradient boosting

Ao invés de fazer apenas um preditor que vai funcionar para a base inteira, ele vai criar várias preditores pequenos e sequenciais. Ele cria um preditor, faz as predições e calcula o erro. O que ele aprende sobre o erro serve de aprendizado para o próximo. É sequencial. E assim por diante. Ele vai criando um modelo que vai funcionar para nossa base.

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gb = GradientBoostingRegressor(random_state= SEED)
```

```python
gb.fit(X_train,y_train)
gb_predict = gb.predict(X_test)
gb_predict
```

```python
gb_predict[gb_predict < 0 ] = 0
```

```python
resultados = pd.DataFrame()
resultados['revenue'] = y_test
resultados['predict'] = gb_predict
resultados['erro'] = gb_predict - y_test
resultados.head()
```

```python
resultados[resultados.revenue > 0]
```

```python
np.sqrt(mean_squared_error(y_test,gb_predict))
```

Temos um modelo razoável para colocar em produção, tivemos um resultado ainda melhor em relação ao anterior. Lembrando do desafio que é conseguir prever quanto um usuário que entra no site vai gastar. Temos diversos tipos de usuários diferentes, é um cenário muito complexo.
